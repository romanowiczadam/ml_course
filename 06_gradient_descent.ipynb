```
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(42)
```

```
X1 = np.array([1,2,3,4,5,6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)
print(X1)
print(Y)

```

```
X1 = X1.reshape(m,1)
Y = Y.reshape(-1,1)
print(X1)
print(Y)
```

```
o = np.ones([m,1])
X_1 = np.append(o, X1, axis = 1)
print(X_1.shape)
X_1
```

```
# wskaźnik uczenia
eta = 0.01

weights = np.random.randn(2, 1)
print(X_1)
# te liczby są naszym poczatkowym w0 i w1 
print(weights)
```

Metoda gradientu prostego

```
# tworzymy dwie listy
# bedziemy do nich dodawac wartosci naszego w0 
intercept = []
# oraz w1
coef = []

# wyznaczamy gradient (wzór z wykładu)
for i in range(3000):
  gradient = (2/m) * X_1.T.dot(X_1.dot(weights) - Y)
  weights = weights - eta * gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])

print(weights)
print(intercept)

```

Wizualizacja dopasowania

```
df = pd.DataFrame(data = {'i' : intercept, 'c': coef})
df
```

```
px.line(df, y = 'i', width=600, title = 'Dopasowanie: intercept')
```

```
px.line(df, y = 'c', width= 600, title = 'Dopasowanie : coef')
```
